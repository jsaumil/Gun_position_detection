```python
import cv2
import os

# Define paths
input_folder = 'data\photos\LMG\project.4'  # Folder containing original images
output_folder = 'final.image'  # Folder to save grayscale images
os.makedirs(output_folder, exist_ok=True)  # Create output folder if it doesn't exist

# Function to convert images to grayscale
def convert_to_grayscale(input_folder, output_folder):
    # Iterate through all files in the input folder
    for filename in os.listdir(input_folder):
        if filename.endswith('.jpg') or filename.endswith('.png'):  # Process only image files
            # Load the image
            image_path = os.path.join(input_folder, filename)
            image = cv2.imread(image_path)
            
            # Convert the image to grayscale
            grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            
            # Save the grayscale image
            output_image_path = os.path.join(output_folder, filename)
            cv2.imwrite(output_image_path, grayscale_image)
            
            print(f"Converted {filename} to grayscale and saved to {output_folder}.")

# Run the grayscale conversion
convert_to_grayscale(input_folder, output_folder)
```
